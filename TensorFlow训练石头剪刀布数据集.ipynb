```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# 1. 确保路径正确（根据之前的解压结构调整）
base_dir = '/content/rps'
if os.path.exists('/content/rps/rps'):  # 检查是否存在二级目录
    base_dir = '/content/rps/rps'

rock_dir = os.path.join(base_dir, 'rock')
paper_dir = os.path.join(base_dir, 'paper')
scissors_dir = os.path.join(base_dir, 'scissors')

# 2. 获取文件列表（按字母顺序排序，确保可重复性）
rock_files = sorted(os.listdir(rock_dir))
paper_files = sorted(os.listdir(paper_dir))
scissors_files = sorted(os.listdir(scissors_dir))

# 3. 显示图片（调整为Colab友好格式）
pic_index = 2  # 从第2张图片开始显示

# 创建子图显示
plt.figure(figsize=(10, 10))

# 显示rock图片
for i in range(2):
    plt.subplot(3, 2, i+1)
    img = mpimg.imread(os.path.join(rock_dir, rock_files[pic_index-2+i]))
    plt.imshow(img)
    plt.title(f"Rock: {rock_files[pic_index-2+i]}")
    plt.axis('off')

# 显示paper图片
for i in range(2):
    plt.subplot(3, 2, i+3)
    img = mpimg.imread(os.path.join(paper_dir, paper_files[pic_index-2+i]))
    plt.imshow(img)
    plt.title(f"Paper: {paper_files[pic_index-2+i]}")
    plt.axis('off')

# 显示scissors图片
for i in range(2):
    plt.subplot(3, 2, i+5)
    img = mpimg.imread(os.path.join(scissors_dir, scissors_files[pic_index-2+i]))
    plt.imshow(img)
    plt.title(f"Scissors: {scissors_files[pic_index-2+i]}")
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1. 确认正确路径
TRAINING_DIR = "/content/rps/rps/"  # 包含rock/paper/scissors子目录
VALIDATION_DIR = "/content/rps-test-set/rps-test-set/"

# 2. 数据增强
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest')

valid_datagen = ImageDataGenerator(rescale=1./255)

# 3. 创建数据生成器
train_generator = train_datagen.flow_from_directory(
    TRAINING_DIR,
    target_size=(150,150),
    batch_size=32,
    class_mode='categorical')

validation_generator = valid_datagen.flow_from_directory(
    VALIDATION_DIR,
    target_size=(150,150),
    batch_size=32,
    class_mode='categorical')

# 4. 模型架构
model = tf.keras.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation='relu', input_shape=(150,150,3)),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Conv2D(128, (3,3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2,2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dropout(0.5),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax')  # 3个输出对应3个类别
])

model.compile(loss='categorical_crossentropy',
              optimizer='rmsprop',
              metrics=['accuracy'])

# 5. 训练
history = model.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=25,
    validation_data=validation_generator,
    validation_steps=len(validation_generator)
)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# 设置图形大小和DPI
plt.figure(figsize=(12, 6), dpi=100)

# 1. 绘制准确率曲线
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], 'r', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], 'b', label='Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim([0, 1])  # 固定y轴范围
plt.legend()

# 2. 绘制损失曲线
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], 'r', label='Training Loss')
plt.plot(history.history['val_loss'], 'b', label='Validation Loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()  # 自动调整子图间距
plt.show()
```
